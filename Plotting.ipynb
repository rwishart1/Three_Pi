```c++
HistMaker hm2("ParticleVars","particleTrees/ParticleVariables_1.root","hists.root");
hm2.AddAxis("ElectronP",2,0,13);
hm2.AddAxis("ElectronSector",7,0,7);
hm2.AddCut("test1","ElectronP>-1");
```

```c++
hm2.ApplyFullSplits();
hm2.ApplySingleSplits();
//hm2.ConfigureFileDirs();
```

```c++
hm2.Histo1D("ElectronTime","Beam Time",100,-5,5);
hm2.Histo1D("MissMass2","Missing Mass",100,-5,5);
hm2.Histo1D("ProtonTime","Proton Time",100,-5,5);
hm2.Histo1D("PipTime","K+ Time",100,-5,5);
hm2.Histo2D("PipTime","PipP","K+ P V Time",50,-5,5,50,0,2);
```

```c++
hm2.SaveHists();
  
```
